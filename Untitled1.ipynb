```python
font = 'Times New Roman'
plt.rcParams['font.size'] = 20
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = font
plt.rcParams['mathtext.it'] = font + ':italic'
plt.rcParams['mathtext.bf'] = font + ':bold'
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
```

```python
ifnum = "01"

obsnums = [
#     "086949_086950",
    "086951_086952",
    "087100_087101",
    "087104_087105",
]
```

```python
paths = [LPRODUCTS / f"T_chbinned_{num}_{ifnum}.nc" for num in obsnums]
T = xr.concat(map(load_netcdf, paths), dim="t")
```

```python
T_chbinned = pl.bin_channels(T, 1)
```

```python
weight = T_chbinned.std("ch") ** -2
spec = (weight * T_chbinned).sum("t") / weight.sum("t")
resampled_freq = np.linspace(spec.ch.min(), spec.ch.max(), 10000)
```

```python
fitter = fitting.LevMarLSQFitter()
model = models.Sine1D(amplitude=0.001, frequency=1.2, phase=0)

ch_use = ~((spec.ch >= 138.2) & (spec.ch <= 138.7))
model_fit = fitter(model, spec.ch, spec, weights=ch_use)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(10, 6), dpi=100)

ax = axes[0]
ax.step(spec.ch, spec, label="Data")
ax.step(spec.ch, model_fit(spec.ch), label="Fit")
ax.set_title(repr(model_fit))
ax.set_ylim(-0.005, 0.014)
ax.set_ylabel("Ta* (K)")
ax.set_xticklabels([])
ax.legend(loc="upper right")

ax = axes[1]
ax.step(spec.ch, spec - model_fit(spec.ch), label="Residual")
ax.set_ylim(-0.005, 0.014)
ax.set_ylabel("Ta* (K)")
ax.set_xticklabels([])
ax.legend(loc="upper right")

ax = axes[2]
ax.step(spec.ch, ch_use, label="Weights")
ax.set_ylim(-0.1, 1.1)
ax.set_xlabel("Observed frequency (GHz)")
ax.legend(loc="upper right")
```

```python
fitter = fitting.LevMarLSQFitter()
model = models.Linear1D(slope=0.0, intercept=0.0)

ch_use = ~((spec.ch >= 138.2) & (spec.ch <= 138.7))
model_fit = fitter(model, spec.ch, spec, weights=ch_use)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(12, 6), dpi=100)

ax = axes[0]
ax.step(spec.ch, spec, label="Data")
ax.step(spec.ch, model_fit(spec.ch), label="Fit")
ax.set_title(repr(model_fit))
ax.set_ylim(-0.005, 0.014)
ax.set_ylabel("Ta* (K)")
ax.set_xticklabels([])
ax.legend(loc="upper right")

ax = axes[1]
ax.step(spec.ch, spec - model_fit(spec.ch), label="Residual")
ax.set_ylim(-0.005, 0.014)
ax.set_ylabel("Ta* (K)")
ax.set_xticklabels([])
ax.legend(loc="upper right")

ax = axes[2]
ax.step(spec.ch, ch_use, label="Weights")
ax.set_ylim(-0.1, 1.1)
ax.set_xlabel("Observed frequency (GHz)")
ax.legend(loc="upper right")
```
